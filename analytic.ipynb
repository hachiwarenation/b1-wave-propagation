```python
import waveprop as wv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
c = 1
M = 2001
N = 31

solver = wv.Solver(c,N,M)
x = solver.get_x()
t = solver.get_t()
T = []

alphas = np.linspace(0.5,1,N)
colors = mpl.colormaps["cool"](np.linspace(0,1,N))

fig,ax = plt.subplots(1,1)
for i in range(N):
    if i % 5 == 0:
        ax.plot(x,wv.wave_g(x-c*t[i]),color=colors[i],alpha=alphas[i])
        T.append(t[i])
ax.grid()
ax.set_xlabel("position x")
ax.set_ylabel("value u")

sm = mpl.cm.ScalarMappable(cmap="cool")
fig.colorbar(sm,ax=ax,label="time t",values=T)
```
